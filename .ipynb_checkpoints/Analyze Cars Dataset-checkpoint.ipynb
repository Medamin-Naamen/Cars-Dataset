Q. 1) Instruction ( For Data Cleaning ) - Find all Null Values in the dataset.
If there is any null value in any column, then fill it with the mean of that column.

```python
data.isnull().sum()
```

```python
data.isnull().any(axis=0)
```

```python
data.fillna(data.mean(numeric_only=True), inplace=True)
```

```python
# Supprimer les dollars pour calculer
data['MSRP']= data['MSRP'].replace('[\$,]','',regex=True).astype(float)
data['Invoice']= data['Invoice'].replace('[\$,]','',regex=True).astype(float)
# filtrer sur place 
data.fillna(data.mean(numeric_only=True), inplace=True)
```

```python
data.isnull().sum()
```
